```


# File location and type
file_location = "/FileStore/tables/nonConfidentialNew.csv"
file_type = "csv"

# CSV options
infer_schema = "false"
first_row_is_header = "True"
delimiter = ","

# The applied options are for CSV files. For other file types, these will be ignored.
df = spark.read.format(file_type) \
  .option("inferSchema", infer_schema) \
  .option("header", first_row_is_header) \
  .option("sep", delimiter) \
  .load(file_location)

display(df)
  
```

```
# File location and type
file_location = "/FileStore/tables/confidential_snappy.parquet"
file_type="parquet"

# CSV options
infer_schema = "false"
first_row_is_header = "True"


# The applied options are for CSV files. For other file types, these will be ignored.
dfparquet = spark.read.format(file_type) \
  .option("inferSchema", infer_schema) \
  .option("header", first_row_is_header) \
  .load(file_location)

display(dfparquet)
```

```
dfparquet.count()
```

```
newdf=df.union(dfparquet)
display(newdf)
```

```
newdf.createOrReplaceTempView("Ftable")
```
